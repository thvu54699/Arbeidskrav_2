```python
# Oppgave 3

import numpy as np

# Funksjon som konverterer grader til radianer
def grader_til_radianer(v_grad):
    v_rad = v_grad * np.pi / 180
    return v_rad

# Hovedprogram
v_grad = float(input('Skriv inn gradtallet: '))
v_rad = grader_til_radianer(v_grad) 
print(f"{v_grad} grader tilsvarer {v_rad:.4f} radianer.")   # Skriver ut radianer med fire desimal

```

```python
# Oppgave 5

import numpy as np

def funk_Areal_Omkrets(a, b):
    r = a / 2
    Areal_Sirkel = np.pi * r**2
    Areal_Trekant = (b * a) / 2
    Omkrets_Sirkel = 2 * np.pi * r
    c = np.sqrt(a**2 + b**2)         # Regner ut hypotenus c
    
    Areal = (Areal_Sirkel / 2) + Areal_Trekant
    Omkrets = (Omkrets_Sirkel / 2) + b + c
    return (Areal, Omkrets)

# Hoved program

a = float(input('Oppgi katet/side a: '))
b = float(input('Oppgi katet/side b: ')) 

(A, O) = funk_Areal_Omkrets(a, b)
print(f"Areal til figur er: {A:.2f} og omkrets er {O:.2f}")   # Skriver ut areal og omkrets med to desimal


```

```python
# Oppgave 6

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)

y = -x**2 - 5     # f(x) = -x**2 - 5

plt.figure(1)
plt.plot(x, y, color="red", linestyle="-", marker=".", linewidth = 3)
plt.grid()
# plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Funksjon f(x) = -x^2 -5")
```
